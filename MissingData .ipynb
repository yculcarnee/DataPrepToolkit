```python
linreg = LinearRegression()
```

```python
dataset
```

```python
data_with_null = dataset[['PassengerId', 'Pclass', 'Survived', 'SibSp', 'Parch', 'Fare', 'Age']]
```

```python
data_with_null.head(15)
```

```python
data_without_null = data_with_null.dropna()
```

```python
data_without_null.head(15)
```

```python
train_data_x = data_without_null.iloc[:,:6] #All features except Age
```

```python
train_data_x.head()
```

```python
train_data_y = data_without_null.iloc[:,6] #Only Age
```

```python
train_data_y.head()
```

```python
linreg.fit(train_data_x, train_data_y)
```

```python
test_data = data_with_null.iloc[:,:6]
age_predicted = pd.DataFrame(columns = ['Age'])
```

```python
age_predicted['Age'] = linreg.predict(test_data)
```

```python
data_with_null['Age'].fillna(age_predicted['Age'], inplace=True)
```
